```python
X_train,X_test,Y_train,Y_test=train_test_split(sales[['TV']],sales[['Sales']],test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
prediction=model.predict(X_test)
print(prediction)
```

```python
model.coef_
```

```python
model.intercept_
```
